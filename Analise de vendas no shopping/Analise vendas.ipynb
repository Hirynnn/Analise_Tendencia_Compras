```python
#importando base de dados
import pandas as pd
import matplotlib as plt 
```

```python
#definindo variavel
df=pd.read_csv('shopping_trends_updated.csv')
```

```python
#Verificando o conjunto de dados
display(df)
```

```python
# Tradução das colunas
traducoes = {
    'Customer ID': 'ID do Cliente',
    'Age': 'Idade',
    'Gender': 'Gênero',
    'Item Purchased': 'Item Comprado',
    'Category': 'Categoria',
    'Purchase Amount (USD)': 'Valor da Compra (USD)',
    'Location': 'Localização',
    'Size': 'Tamanho',
    'Color': 'Cor',
    'Season': 'Estação',
    'Review Rating': 'Avaliação',
    'Subscription Status': 'Status de Assinatura',
    'Shipping Type': 'Tipo de Envio',
    'Discount Applied': 'Desconto Aplicado',
    'Promo Code Used': 'Código Promocional Utilizado',
    'Previous Purchases': 'Compras Anteriores',
    'Payment Method': 'Método de Pagamento',
    'Frequency of Purchases': 'Frequência de Compras'
}

# Renomeie as colunas
df = df.rename(columns=traducoes)
```

```python
#Visualizando
df
```

```python
#Verificando os tipos 
df.dtypes
```

```python
# Verificando se há linhas vazias
pd.isnull(df).sum()
```

```python
#Verificando a categoria mais vendida

contagem_categorias = df['Categoria'].value_counts()
contagem_categorias


```

```python
#itens mais vendidos total


contagem_itens=df['Item Comprado'].value_counts()
contagem_itens
```

```python
# Média de avaliação por item
aval_itens=df.groupby('Item Comprado')['Avaliação'].mean()
aval_itens
```

```python
# Itens mais vendidos por estação 
itens_mais_comprados_por_estacao = df.groupby('Estação')['Item Comprado'].value_counts().groupby(level=0).idxmax()
itens_mais_comprados_por_estacao
```

```python
#calculando o lucro por estação 

lucro = df.groupby('Estação')['Valor da Compra (USD)'].sum()
lucro
```

```python
#Calculando o lucro por cidade 



soma_por_cidade = df.groupby('Localização')['Valor da Compra (USD)'].sum()
soma_por_cidade = soma_por_cidade.sort_values(ascending=False)
soma_por_cidade.plot(kind='bar', figsize=(12, 6))

plt.xlabel('Localização')
plt.ylabel('Soma do Valor da Compra (USD)')
plt.title('Soma do Valor da Compra por Cidade (Ordem Decrescente)')

plt.show()
```

```python
#calculando a quantidade de assinantes

assinantes=df['Status de Assinatura'].value_counts()
assinantes

```

```python
# Idade média dos clientes 
df.Idade.mean()
```

```python
# Metodo de pagamento mais frequente
metodo_pagamento=df['Método de Pagamento'].value_counts()
metodo_pagamento
```

```python
# Metodo de entrega mais frequente 
tipo_entrega=df['Tipo de Envio'].value_counts()
tipo_entrega
```

```python
# Diferença de gênero por quantidade de clientes
genero_compra=df['Gênero'].value_counts()
genero_compra
```

```python
#Frequencia de compras por id cliente

frequencia = df.groupby('Frequência de Compras')['ID do Cliente'].nunique()
frequencia
```

```python
# Quantidade de compras com desconto

desconto= df['Desconto Aplicado'].value_counts()
desconto

```

# Resultado de análise

O resultado da análise revela um conjunto de dados notavelmente bem distribuído, apresentando uma distribuição equilibrada de informações. No entanto, apesar dessa uniformidade, é possível encontrar algumas discrepâncias que merecem atenção.

Uma observação notável é que as categorias 'Clothing' e 'Accessories' representam mais de 50% do lucro total de vendas, independente da estação do ano. Isso sugere que essas categorias desempenham um papel fundamental nas receitas da empresa.

Além disso, chama a atenção o fato de que 73% das compras são feitas por não assinantes, indicando uma oportunidade de aumentar a base de assinantes para impulsionar o negócio.

A idade média dos compradores é de 44 anos, sugerindo que a maioria dos clientes pertence a faixas etárias mais maduras. Essa informação pode ser valiosa para adaptar estratégias de marketing e produtos para atender às preferências desse público.

Outro ponto de destaque é que 68% do público é do sexo masculino, o que pode influenciar a seleção de produtos e campanhas de marketing.

Finalmente, 57% das compras não fazem uso de cupons, o que pode indicar que existe espaço para implementar estratégias de descontos e promoções para atrair mais compradores.

Esses insights, embora desafiadores de se obter devido à distribuição uniforme dos dados, proporcionam uma visão valiosa para a tomada de decisões estratégicas e o aprimoramento do desempenho da empresa.
